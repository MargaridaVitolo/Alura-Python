```python
import pandas as pd
```

```python
df = pd.read_csv ("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
novo_nomes_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=novo_nomes_colunas)
display(df.columns)
```

```python
df_limpo = df.dropna()
```

```python
df.info()
```

```python
df_limpo.info()
```

```python
df_limpo.head()
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df_limpo['remoto'] = df_limpo['remoto'].replace(traducao_remoto)
display(df_limpo['remoto'].value_counts())
```

```python
traducao_senioridade = {
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Sênior',
    'EX': 'Executivo'
}

df_limpo['senioridade'] = df_limpo['senioridade'].replace(traducao_senioridade)
display(df_limpo['senioridade'].value_counts())
```

```python
traducao_executivo = {
    'Estagiário' : 'Executivo'
}

df_limpo['senioridade'] = df_limpo['senioridade'].replace(traducao_executivo)
display(df_limpo['senioridade'].value_counts())
```

```python
df_limpo = df_limpo.assign(ano = df_limpo['ano'].astype('int64'))
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
