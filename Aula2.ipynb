```python
novo_nomes_colunas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=novo_nomes_colunas)
display(df.columns)
```

```python
import numpy as np
#CriaÃ§Ã£o de um dataframe de teste
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
    })
#calcula mÃ©dia e substitui nulos e arredonda valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
'''ComentÃ¡rio
em mais de uma linha
'''
#calcula mediana

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})
#ffill completa com o valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'TerÃ§a', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})
#bfill completa com o valor seguinte
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas

```

```python
df_cidades = pd.DataFrame({
   'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
   'cidade': ['SÃ£o Paulo', np.nan, 'Curitiba', np.nan, 'Rio de Janeiro']
})
# preenche com valores especÃ­ficos
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('NÃ£o informado')
df_cidades
```
